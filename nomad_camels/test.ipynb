```python
import numpy as np
import matplotlib.pyplot as plt

def fun(x, a, b, c):
    return a * (np.exp(b * x) -1 )
```

```python
ran = np.linspace(-1, 1, 100)

a = 1e-5
b = 10
c = -1

y = fun(ran, a, b, c)

plt.plot(ran, np.abs(y))
plt.semilogy()
plt.tight_layout()
[0.00000000e+00 2.90000000e+02 2.00000000e+01 5.00000000e-01
 2.13665460e-03 1.84936693e+00]
```
